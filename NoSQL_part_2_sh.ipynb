### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
first_result = establishments.find_one(query)
print("First document in the results:")
pprint(first_result)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Per instructions, "coerce non-numeric RatingValue to null"
establishments.update_many(
    {"RatingValue": {"$exists": True, "$type": "string"}},
    [{"$set": {"RatingValue": {"$cond": [{"$regexMatch": {"input": "$RatingValue", "regex": "^[0-9]+$"}}, {"$toInt": "$RatingValue"}, None]}}}]
)

# Verify the changes
sample_documents = establishments.find({"RatingValue": {"$ne": None}}).limit(10)
print("Sample documents after conversion:")
for doc in sample_documents:
    pprint(doc)

```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "LocalAuthorityName": "London",
    "RatingValue": {"$gte": "4"}
}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments with London as the Local Authority and RatingValue >= 4: {count}")

# Display the first document in the results using pprint
first_result = establishments.find_one(query)
print("First document in the results:")
pprint(first_result)
```

```python
# Fetch distinct values of LocalAuthorityName
distinct_authorities = establishments.distinct("LocalAuthorityName")

# Print the distinct values
print("Distinct LocalAuthorityName values:")
for authority in distinct_authorities:
    print(authority)
```

```python
# using hint "The London Local Authority has a longer name - use $regex"
# use query to find "City of London Corporation"
# Define the query to find establishments with LocalAuthorityName containing "London" and a RatingValue >= 4
query = {
    "LocalAuthorityName": {"$regex": "London", "$options": "i"},
    "RatingValue": {"$gte": 4}
}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments with LocalAuthorityName containing 'London' and RatingValue >= 4: {count}")

# Display the first document in the results using pprint
first_result = establishments.find_one(query)
print("First document in the results:")
pprint(first_result)
```

```python
# Convert the result to a Pandas DataFrame
results = establishments.find(query)
df = pd.DataFrame(list(results))

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df)}")
# Display the first 10 rows of the DataFrame
print("First 10 rows in the DataFrame:")
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Convert latitude and longitude to float if they are not already
establishments.update_many(
    {"geocode.latitude": {"$type": "string"}},
    [{"$set": {"geocode.latitude": {"$toDouble": "$geocode.latitude"}}}]
)

establishments.update_many(
    {"geocode.longitude": {"$type": "string"}},
    [{"$set": {"geocode.longitude": {"$toDouble": "$geocode.longitude"}}}]
)
```

```python
# Latitude and longitude of "Penang Flavours"
latitude = 51.49014200
longitude = 0.08384000

# Define the degree search range
degree_search = 0.01

# Define the query to find establishments near the specified location within the degree range
query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}

# Define the sort order (by hygiene score in ascending order)
sort = [("scores.Hygiene", 1)]

# Limit the results to the top 5
limit = 5

# Execute the query with sorting and limiting and convert to a list
results_list = list(establishments.find(query).sort(sort).limit(limit))

# Print the results using pprint
print("Top 5 establishments with a RatingValue of 5, sorted by lowest hygiene score, nearest to 'Penang Flavours':")
for result in results_list:
    pprint(result)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results_list)

# Print the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of documents in the result: {len(results)}")

# Print the first 10 results
print("First 10 results:")
for result in results[:10]:
    pprint(result)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)

# Rename the columns for better readability
df.rename(columns={'_id': 'LocalAuthorityName', 'count': 'EstablishmentsWithHygieneScore0'}, inplace=True)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
df.head(10)
```
